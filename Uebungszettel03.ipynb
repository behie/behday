```python
import math
def pearson(x,y):

    x_m = (sum(x)/len(x))
    y_m = (sum(y)/len(y))

    s1 = 0.0
    s2 = 0.0
    s3 = 0.0

    for i in range (len(x)):
        s1 += (x[i]-x_m)*(y[i]-y_m)
        s2 += (x[i]-x_m)*(x[i]-x_m)
        s3 += (y[i]-y_m)*(y[i]-y_m)

    s4 = math.sqrt(s2*s3)

    r = s1/s4
    return(r)
```

```python
import pandas as pd
import math

def pearson(x,y):
    
    x_m = float(sum(x))/len(x)
    y_m = float(sum(y))/len(y)

    s1 = 0.0
    s2 = 0.0
    s3 = 0.0

    for i in range (len(x)):
        s1 += (x[i]-x_m)*(y[i]-y_m)
        s2 += (x[i]-x_m)*(x[i]-x_m)
        s3 += (y[i]-y_m)*(y[i]-y_m)

    s4 = math.sqrt(s2*s3)

    r = s1/s4
    return(r)


df = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')

dfg = df[df["country"] == "Germany"]
a = dfg["year"].tolist()
b = dfg["co2"].tolist()
print ("Germany:",pearson(a,b))

dfw = df[df["country"] == "World"]
c = dfw["year"].tolist()
d = dfw["co2"].tolist()
print ("\nWorld:", pearson(c,d))
```
